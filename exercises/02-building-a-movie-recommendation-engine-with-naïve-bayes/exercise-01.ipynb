## 1. Preparing the data

```python
import pandas as pd
```

- Load the data files:

```python
ratings_file_path = '../../data/ml-1m/ratings.dat'
movies_file_path = '../../data/ml-1m/movies.dat'
users_file_path = '../../data/ml-1m/users.dat'
```

```python
ratings_df = pd.read_csv(ratings_file_path, sep='::', header=None, engine='python')
ratings_df.columns = ['UserID', 'MovieID', 'Rating', 'Timestamp']
ratings_df
```

```python
users_df = pd.read_csv(users_file_path, sep='::', header=None, engine='python')
users_df.columns = ['UserID', 'Gender', 'Age', 'Occupation', 'Zip-code']
users_df
```

```python
movies_df = pd.read_csv(movies_file_path, sep='::', header=None, engine='python')
movies_df.columns = ['MovieID', 'Title', 'Genres']
movies_df
```

- View summary of the loaded DataFrames:

```python
ratings_df.info()
```

```python
users_df.info()
```

```python
movies_df.info()
```

- Combine the DataFrames:

```python
# Check for unique values
print(
  f"No. of unique movie IDs: {movies_df['MovieID'].nunique()}",
  f"\nNo. of unique movie titles: {movies_df['Title'].nunique()}"
)
```

```python
# Create a mapping dictionary using movies_df['MovieID'] and movies_df['Title']
movieID_title = dict(
  zip(movies_df['MovieID'], movies_df['Title'])
)

# View the first 10 items in movieID_title
count = 0

for key, value in movieID_title.items():
  if count < 10:
    print(f"{key}: {value}")
    count += 1
  else:
    break
```

```python
# Duplicate ratings_df to create user_movie_ratings
user_movie_ratings = ratings_df.copy()

# Create a new column using .map()
user_movie_ratings['Title'] = user_movie_ratings['MovieID'].map(movieID_title)

user_movie_ratings.head()
```

```python
# Add movie genres to user_movie_ratings
title_genre = dict(
  zip(movies_df['Title'], movies_df['Genres'])
)

count = 0

for key, value in title_genre.items():
  if count < 10:
    print(f"{key}: {value}")
    count += 1
  else:
    break
```

```python
user_movie_ratings['Genres'] = user_movie_ratings['Title'].map(title_genre)

user_movie_ratings.head()
```

```python
# Add gender of users to user_movie_ratings
userID_gender = dict(
  zip(users_df['UserID'], users_df['Gender'])
)

count = 0

for key, value in userID_gender.items():
  if count < 10:
    print(f"{key}: {value}")
    count += 1
  else:
    break
```

```python
user_movie_ratings['Gender'] = user_movie_ratings['UserID'].map(userID_gender)

user_movie_ratings.head()
```

```python
# Add age of users to user_movie_ratings
userID_age = dict(
  zip(users_df['UserID'], users_df['Age'])
)

count = 0

for key, value in userID_age.items():
  if count < 10:
    print(f"{key}: {value}")
    count += 1
  else:
    break
```

```python
user_movie_ratings['Age'] = user_movie_ratings['UserID'].map(userID_age)

user_movie_ratings.head()
```

```python
# Add occupation of users to user_movie_ratings
userID_occupation = dict(
  zip(users_df['UserID'], users_df['Occupation'])
)

count = 0

for key, value in userID_occupation.items():
  if count < 10:
    print(f"{key}: {value}")
    count += 1
  else:
    break
```

```python
user_movie_ratings['Occupation'] = user_movie_ratings['UserID'].map(userID_occupation)

user_movie_ratings.head()
```

```python
# Add zip code of users to user_movie_ratings
userID_zipcode = dict(
  zip(users_df['UserID'], users_df['Zip-code'])
)

count = 0

for key, value in userID_zipcode.items():
  if count < 10:
    print(f"{key}: {value}")
    count += 1
  else:
    break
```

```python
user_movie_ratings['Zip-code'] = user_movie_ratings['UserID'].map(userID_zipcode)

user_movie_ratings.head()
```

- View a concise summary of the combined DataFrame:

```python
user_movie_ratings.info()
```

- Rearrange the order of columns in the DataFrame:

```python
column_order = ['UserID', 'Gender', 'Age', 'Occupation', 'Zip-code', 'MovieID', 'Title', 'Genres', 'Timestamp', 'Rating']

user_movie_ratings = user_movie_ratings[column_order]

user_movie_ratings.head()
```

- Change the `dtype` in each column:

```python
# Change the dtype in the gender column from object to category (unordered)
user_movie_ratings['Gender'] = user_movie_ratings['Gender'].astype('category')

print(f"Gender column Dtype: {user_movie_ratings['Gender'].dtype}")
print(f"Gender column categories: {user_movie_ratings['Gender'].cat.categories}")
```

```python
# Change the dtype in the age column from int64 to category (ordered)
bins = [0, 12, 19, 34, 54, 64, 120]
# labels = ['Child', 'Teen', 'Young Adult', 'Middle Aged', 'Older Adult', 'Senior']
labels = ['Child (0 - 12)', 'Teen (13 - 19)', 'Young Adult (20 - 34)', 'Middle Aged (35 - 54)', 'Older Adult (55 - 64)', 'Senior (65+)']

user_movie_ratings['Age Group'] = pd.cut(
  user_movie_ratings['Age'],
  bins=bins,
  labels=labels,
  include_lowest=True
)

user_movie_ratings.head()
```

```python
age_column = user_movie_ratings.pop('Age')

column_order = ['UserID', 'Gender', 'Age Group', 'Occupation', 'Zip-code', 'MovieID', 'Title', 'Genres', 'Timestamp', 'Rating']
user_movie_ratings = user_movie_ratings[column_order]

print(user_movie_ratings.head())
print("-" * 50)
print(f"\nAge Group column Dtype: {user_movie_ratings['Age Group'].dtype}")
print(f"Age Group column categories: {user_movie_ratings['Age Group'].cat.categories}")
print(f"Are the Age Group column categories ordered? {user_movie_ratings['Age Group'].cat.ordered}")
```
